```python
age=int(input("How old are you?"))
if age>=10 and age<=20:
  print("You are allowed in the the class")
else:
  grade=int(input("What grade are you?"))
  if grade==10:
    print(f"Your age {age} fits this class")
  elif grade==11:
    print("You are not allowed to enroll")
  elif grade==12:
    print("You are not allowed to enroll")
  else:
    print("You can not enroll")

```

```python
alphabet=input("Please enter any character:")
if alphabet>="A" and alphabet<="Z":
  print("This character is an alphabet and is uppercase")
elif alphabet>="a" and alphabet<="z":
  print("This character is an alphabet and is lowercase")
else:
  print("This is not an alphabet")

```
